Import Dependencies

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Preparing Dataset

```python
x_data = []
y_data = []
with open('trainingData/collectedImages.data', 'r') as openedFile:
    for line in openedFile:
        if len(line) < 1:
            continue
        else:
            lineAsList = []
            for i in range(len(line) - 1):
                lineAsList.append(int(line[i]))
            x_data.append(lineAsList)
with open('trainingData/collectedLabels.data', 'r') as openedFile:
    for line in openedFile:
        if len(line) < 1:
            continue
        else:
            number = int(line)
            results = [0]*10
            results[number] = 1
            y_data.append(number)
```

```python
x_train = x_data[0:int(len(x_data)*0.9)]
y_train = y_data[0:int(len(y_data)*0.9)]

x_test = x_data[int(len(x_data)*0.9) + 1:]
y_test = y_data[int(len(y_data)*0.9) + 1:]
```

```python
print(x_test)
```

```python
input_shape = (len(x_data[0]),)
print(input_shape)
```

Setting up the Model

```python
model = Sequential()
model.add(Dense(28*28, activation='relu', input_shape=input_shape))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Training the Model

```python
history = model.fit(x_train, y_train, epochs=10, verbose=1, validation_data=(x_test, y_test))
```

```python
model.evaluate(x_test, y_test)
```

Saving the Model

```python
model.save('trainedModel.h5')
```
